```python
dict_inds = {i: [] for i in range(len(classes))}
dict_negs = {i: [] for i in range(len(classes))}
    
for idd, _y in enumerate(y_train):
    if len(_y) == 1:
        dict_inds[_y[0]].append(idd)
for idd, _y in enumerate(y_train):
    for j in _y:
        dict_negs[j].append(idd)

xtrlen = len(x_train)
```

```python
def create_trips(size):
    
    
    anchs = np.zeros((size, 224, 224, 3))
    posis = np.zeros((size, 224, 224, 3))
    negs = np.zeros((size, 224, 224, 3))
    
    randinds = random.sample(range(xtrlen), size)
    for iid, ind in enumerate(randinds):
        anchs[iid] = x_train[ind]
        one_ans = random.sample(y_train[ind], 1)[0]
        posis[iid] = x_train[random.sample(dict_inds[one_ans], 1)]
        one_neg = random.sample([e for e in range(len(classes)) if e not in y_train[ind]], 1)[0]
        negs[iid] = x_train[random.sample(dict_negs[one_neg], 1)]
    return [anchs, posis, negs]
        
```

## model design

```python
def triplet_loss(y_true, y_pred):
    margin = 0.05
    # y_true is not used since there are no true values
    anchor, positive, negative = y_pred[0], y_pred[1], y_pred[2]
    
    # Calculate the distance between anchor and positive
    pos_distance = tf.reduce_sum(tf.square(anchor - positive), axis=-1)

    # Calculate the distance between anchor and negative
    neg_distance = tf.reduce_sum(tf.square(anchor - negative), axis=-1)

    # Compute the loss with a margin
    loss = tf.reduce_mean(tf.maximum(0.0, pos_distance - neg_distance + margin))

    return loss
```

```python
inp1, inp2, inp3 = tf.keras.layers.Input(shape=(224,224,3)), tf.keras.layers.Input(shape=(224,224,3)), tf.keras.layers.Input(shape=(224,224,3))
feat1, feat2, feat3 = feat_extract(inp1), feat_extract(inp2), feat_extract(inp3)
```
